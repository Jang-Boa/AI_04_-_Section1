# 데이터 전처리 작업
-----
1. 'Year' 에서 연도를 동일하게 전부 나오도록 표시하고 int형으로 변환
2. 각 지역 sales 마다 존대하는 "M", "K" 를 숫자로 변환시키고 데이터를 float형으로 변환
3. feature 상관없이 데이터에 존재하는 결측치는 제거하여 분석을 진행
4. 전 지역의 출고량을 합산하여 전 지역의 전체 출고량을 계산하고 Total_Sales 에 저장

전처리 전 데이터
스크린샷 2021-06-01 오후 10.31.18.png

전처리 후 데이터
스크린샷 2021-06-01 오후 10.32.17.png

```python
before = 16598
after = 16241
percentage_NA = (before - after)/before
if percentage_NA < 10:
  print("결측치는 전체 데이터에서 %.3f%%를 차지하며, 결측치는 전체 데이터의 10%% 미만이므로 제거하기로 했습니다."%(percentage_NA*100))
else:
  print("결측치는 전체 데이터에서 %.3f%%를 차지합니다."%(percentage_NA*100))
```
